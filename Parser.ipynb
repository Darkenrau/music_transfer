```python
playlist = {}




for elem in data['tracks']['data']:
    artist = elem['artist']['name']
    track = elem['title']
       
    if artist not in playlist:
        playlist[artist] = []
    if track in playlist[artist]:
        continue
    
        
    else:
        playlist[artist].append(track)
print(playlist)
```
